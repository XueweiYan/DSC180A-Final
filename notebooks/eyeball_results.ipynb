```python
import pandas as pd
import numpy as np
```

```python
file = "models/DBLP.5K/AutoPhrase.txt"
```

```python
with open(file) as f:
    txt = f.readlines()
```

```python
scores = [float(x.split()[0]) for x in txt]
```

```python
np.sum([s > 0.5 for s in scores])
```

```python
np.sum([s > 0.9 for s in scores])
```

```python
np.sum([((s < 0.55) and (s > 0.5)) for s in scores])
```

```python
np.sum([' ' in x for x in txt])
```

```python
np.sum([not ' ' in x for x in txt])
```
